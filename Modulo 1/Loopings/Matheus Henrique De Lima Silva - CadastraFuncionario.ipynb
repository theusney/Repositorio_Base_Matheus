```python
nomes = ["Matheus" "Lucas" "Juliana" "Pietro"]
cargos = ["chefe" "gerente" "Socia" "jovem aprendiz"]
salarios = ["100000" "30000" "60000" "750"]

```

Para cada funcion√°rio, coletar:

### üéØ Exemplo de Sa√≠da

```python
# Listas para armazenar os dados
nomes = []
cargos = []
salarios = []
quantidade = int(input("Quantos funcion√°rios deseja cadastrar? "))

for i in range(quantidade):
    nome = input("Digite seu nome")
    cargo = input("Digite seu cargo")
    salario = input("digite seu salario")
    
    nomes.append(nome) 
    cargos.append(cargo)
    salarios.append(salario)
    print(nomes, cargos, salarios)

    







 

```
